# Make prediction fun


```python
def prediction(skills_list):
    col_list = joblib.load(DATA_COL)
    pred_df = pd.DataFrame(col_list.isin(skills_list)).T
    pred_df = pred_df.astype(int)
    pred_df.columns = col_list
    pred_df['experanse_years'] = 1

    pred_df_scl = StandardScaler().fit_transform(pred_df.T)

    model = joblib.load(final_model_path)

    result = model.predict(pred_df_scl.T)

    return result[0]
```

```python
def prediction_prob(skills_list):
    col_list = joblib.load(DATA_COL)
    pred_df = pd.DataFrame(col_list.isin(skills_list)).T
    pred_df = pred_df.astype(int)
    pred_df.columns = X.columns
    pred_df['experanse_years'] = 1

    pred_df_scl = StandardScaler().fit_transform(pred_df.T)

    model = joblib.load(final_model_path)

    class_names = joblib.load(Classes_NAME)
    probabilities = model.predict_proba(pred_df_scl.T)
    class_prob_dict = dict(zip(class_names, probabilities.flatten()))
    # Create a DataFrame from the dictionary
    df = pd.DataFrame(list(class_prob_dict.items()),
                      columns=['Class', 'Probability'])

    # Sort the DataFrame by the 'Probability' column in ascending order
    df_sorted = df.sort_values(by='Probability', ascending=False)

    return df_sorted
```

```python
def find_matching_skill_groups(skill_list, skill_dict):
    matching_skill_groups = []

    for group_name, group_skills in skill_dict.items():
        common_skills = set(skill_list) & set(group_skills)
        if len(common_skills) >= 3:
            non_common_skills = list(set(group_skills) - set(common_skills))
            matching_skill_groups.append((group_name, non_common_skills))

    return matching_skill_groups


# Your list of skills
skills_list = ['R', 'Python', 'TensorFlow',
               'Jupyter Notebook/JupyterLab', 'pandas', 'Numpy']

# Your dictionary of skill groups
skills_dict = skills_clusters

# Find matching skill groups
matching_groups = find_matching_skill_groups(skills_list, skills_dict)

# Display matching skill groups and list without common skills
if matching_groups:
    print("Matching skill groups:")
    for group_name, non_common_skills in matching_groups:
        print(f"Group: {group_name}")
        print(f"Original Skills: {skills_dict[group_name]}")
        print(f"Non-Common Skills: {non_common_skills}")
        print("---")
else:
    print("No matching skill groups found.")
```
